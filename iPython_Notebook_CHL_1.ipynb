```python
#LASSO REGRESSION
predvar = df_reduced[['DURBRST', 'GAGEBRTH', 'BIRTHWT', 'BIRTHLEN', 'NCHLD', 
                    'BMI', 'HCIRCM', 'HCT', 'FERRITIN', 'L_MWFBCC', 'MWFCNGLM', 'MWFCRADT',
                    'MWFCRBLM', 'MWFFRNTL', 'MWFGCC', 'MWFICAPS', 'MWFOCCPT', 'MWFORAD',
                   'MWFPARTL', 'MWFSCC', 'MWFSLFAS', 'MWFTMPRL', 'SEXN_1',
                   'SEXN_2', 'FEEDINGN_3.0', 'FEEDINGN_7.34626865672', 'FEEDINGN_9.0', 
                   'FEEDINGN_10.0', 'MMARITN_1.0', 'MMARITN_2.0', 'MMARITN_2.06830122592',
                   'MMARITN_3.0', 'MMARITN_4.0', 'MMARITN_6.0', 'HOLLINGS_0.0',
                   'HOLLINGS_4.0', 'HOLLINGS_5.0', 'HOLLINGS_5.5587431694', 'HOLLINGS_6.0',
                   'HOLLINGS_7.0']]
```

```python
target = df_reduced['MSELC']
```

```python
pred_train, pred_test, tar_train, tar_test = train_test_split(predvar, target, 
                                                             test_size=.2, random_state=123)
```

```python
from sklearn.linear_model import LassoLarsCV
model= LassoLarsCV(cv=10, precompute=False).fit(pred_train,tar_train)
```

```python
dict(zip(predvar.columns, model.coef_))
```

```python
m_log_alphas = -np.log10(model.alphas_)
ax = plt.gca()
plt.plot(m_log_alphas, model.coef_path_.T)
plt.axvline(-np.log10(model.alpha_), color='k',
           label='alpha CV')
plt.ylabel('Regression Coefficients')
plt.xlabel('-log(alpha)')
plt.title('Regression Coefficients Progression for Lasso Paths')
```

```python
m_log_alphascv = -np.log10(model.cv_alphas_)
plt.figure()
plt.plot(m_log_alphascv, model.cv_mse_path_, ':')
plt.plot(m_log_alphascv, model.cv_mse_path_.mean(axis=-1), 'k',
        label='Average across the folds', linewidth=2)
plt.axvline(-np.log10(model.alpha_), color='k',
           label='alpha CV')
plt.legend()
plt.xlabel('-log(alpha)')
plt.ylabel('Mean squared error')
plt.title('Mean squared error on each fold')
```

```python
from sklearn.metrics import mean_squared_error
train_error = mean_squared_error(tar_train, model.predict(pred_train))
test_error = mean_squared_error(tar_test, model.predict(pred_test))
print ('training data MSE')
print(train_error)
print ('test data MSE')
print(test_error)
```

```python
rsquared_train=model.score(pred_train,tar_train)
rsquared_test=model.score(pred_test,tar_test)
print ('training data R-square')
print(rsquared_train)
print ('test data R-square')
print(rsquared_test)
```
